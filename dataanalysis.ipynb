### Question 4

Please group the dataset by `TermDays` and calculate the `mean` loan amount and the `standard deviation` of loan amount within each term.

What is the mean loan amount in 30-day loans and what is the standard deviation of  loan amount in 90-day loans? Please give your answers to the nearest two decimal places. 

```python
listOfDays = set(df['TermDays'])  #getting the unique values in the TermDays column
listOfDays
```

```python
df1 = df.copy()
```

```python

# calculating the mean loan amount and the standard deviation of loan amount within each term

import statistics

allmeans = []
allSDs = []

for i in listOfDays:
    df_0 = df1[df1['TermDays'] == i]    #looping through and intitiating a distinct dataframe for each iteration
    average = statistics.mean(df_0['LoanAmount'])    #calculating the average and SD in each df iteration
    SD = statistics.stdev(df_0['LoanAmount'])
    allmeans.append(round(average, 2))    #rounding the result and appending it to a list
    allSDs.append(round(SD, 2))
    
print(allmeans)
print(allSDs)
```

### Question 5

`Late` loans are defined as loans that were paid `more than 3 days` after their due date. Please find the average number of days that these loans were paid beyond their due date for each `week of the year`. This will require you to break all `appilcationDate's` down into the week of the year they were made and then perform a group by. 

What is the average number of days that `late` loans applied for in the second week of the year were paid late? Please sumbit your answer rounded to two decimal places. 

```python
# week of year and average number of late days

df.head()
```

```python
df['repaymentDueDate'] = pd.to_datetime(df['repaymentDueDate'])
df['repaymentPaidDate'] = pd.to_datetime(df['repaymentPaidDate'])

df['dateDifference'] = df['repaymentDueDate'] - df['repaymentPaidDate']
```

```python
df_conv2 = df.copy()
df_conv2['appilcationDate'] = pd.to_datetime(df_conv2['appilcationDate'])

# Assigning the week of the year values to another column in the dataframe
df_conv2['applicationWeek'] = df_conv2['appilcationDate'].dt.week
        
df_conv2.head()
```

```python
weekNumber = set(df_conv2['applicationWeek'])
df_conv4 = df_conv2.copy()
df_conv4.head()
```

```python


week = []

def eachWeek(): 
    for i in weekNumber:
        df_conv0 = df_conv4[df_conv4['applicationWeek'] == i]
        total = 0
        count = 0
        for j in df_conv0['dateDifference']:
            if j.days < -3:
                total += j.days
                count += 1
                
        if count == 0:   #to cover for zeroDivisionError
            week.append(0)
        else:
            week.append(round(total/count, 2))
        
eachWeek()
```

```python
# Note: Averages with value of 0 means that there are no loans paid less than 3 days late in such week

week
```
